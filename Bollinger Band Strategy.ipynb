```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
def get_sma(prices, rate):
    return prices.rolling(rate).mean()
```

```python
def bollinger_bands_upper(prices, rate=20):
    sma = get_sma(prices, rate)
    std = prices.rolling(rate).std()
    bollinger_up = sma + std * 2 
    bollinger_down = sma - std * 2 
    return bollinger_up

def bollinger_bands_lower(prices, rate=20):
    sma = get_sma(prices, rate)
    std = prices.rolling(rate).std()
    bollinger_up = sma + std * 2 
    bollinger_down = sma - std * 2 
    return bollinger_down
```

```python
start = dt.datetime.now() - dt.timedelta(days=365 * 20)
end = dt.datetime.now()
data = pdr.DataReader('^GSPC', 'yahoo', start, end)
data['BB upper'] = bollinger_bands_upper(data['Adj Close'])
data['BB lower'] = bollinger_bands_lower(data['Adj Close'])
print(data)
plt.figure(figsize=(16,9))
plt.plot(data['Open'], label='Price', color='black')
plt.plot(data['BB upper'], label='BB upper', color='red')
plt.plot(data['BB lower'], label='BB lower', color='green')
plt.show()
```

```python
def buy_sell(data):
    buy = []
    sell = []
    flag = 1
    
    for i in range(len(data)):
        if data['Open'][i] < data['BB lower'][i]:
            if flag == 1:
                buy.append(data['Adj Close'][i])
                sell.append(np.nan)
                #flag = 1
            else:
                buy.append(np.nan)
                sell.append(np.nan)
        elif data['Open'][i] > data['BB upper'][i]:
            if flag == 1:
                buy.append(np.nan)
                sell.append(data['Adj Close'][i])
            else:
                buy.append(np.nan)
                sell.append(np.nan)
        else:
            buy.append(np.nan)
            sell.append(np.nan)
    return buy, sell
```

```python
buy_sell = buy_sell(data)
data['buy signal price'] = buy_sell[0]
data['sell signal price'] = buy_sell[1]
```
